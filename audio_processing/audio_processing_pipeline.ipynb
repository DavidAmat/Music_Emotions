```python
def ls_S3(folder_path, S3_BUCKET = 'musicemotions', maxkeys = 1000):
    #Connect to S3
    s3 = boto3.client("s3")
    
    # S3 list objects
    response = s3.list_objects_v2(
                Bucket=S3_BUCKET,
                Prefix =folder_path,
                MaxKeys=maxkeys )

    files_inside_folder = list()
    for contents_folder in response["Contents"]:
        
        # Get the contents of the folder
        file_names = contents_folder["Key"].split("/")[-1]
        
        #If the name of the file is not empty:
        if len(file_names):
            files_inside_folder.append(file_names)
    return files_inside_folder

def download_mp3(path_song_S3,song_name):
    region_S3 = 'eu-west-1' #specify the region of the S3 bucket, especially if executing this code in a different region than Ireland
    s3 = boto3.client("s3",region_S3)
    
    # Save it locally in the tmp folder as song.mp3
    with open('tmp/song.mp3', 'wb') as f:
        #s3.download_fileobj('musicemotions', os.path.join(path_song_S3,song_name), f)
        s3.download_fileobj('musicemotions',f'{path_song_S3}/{song_name}', f)
    return 

def get_sampling_rate_wav(location_wav_file = "song.wav"):
    comando_sample_rate = f'ffprobe -show_streams {location_wav_file}'
    comando_output = subprocess.check_output(comando_sample_rate, shell=True)
    outs = [line.decode("utf-8")  for line in comando_output.splitlines()]
    for oo in outs:
        if "sample_rate" in oo:
            fs_wav = oo.split("sample_rate=")
            fs_wav = int(fs_wav[1])
            break
    return fs_wav
```

```python
#Sampling Rate
fs_wav = get_sampling_rate_wav()
```

```python
# Load wav
x , sr = librosa.load("song.wav", sr = fs_wav)
```

```python
C = librosa.cqt(x, sr = sr, hop_length=512, window=blackmanharris, n_bins = 84)
```

```python
freqs_cqt = librosa.cqt_frequencies(n_bins = 84, fmin=librosa.note_to_hz('C1'))
```

```python
freqs_cqt = np.round(freqs_cqt)
```

```python
# Spectral Power
SP = abs(C)
SP_norm = librosa.util.normalize(SP)
df = pd.DataFrame(SP_norm, index=freqs_cqt)

# Flipping coordinates
SP_norm_flip = df.sort_index(ascending=False).values
```

```python
# .wav signal
x_dim = x.shape[0] #samples
duration = x_dim / sr # in seconds

# C
C_num_samples = C.shape[1]
img_duration = 30 
C_sampl_sep = duration * 512 / x_dim # in seconds
C_samples_duration_window =  int(np.ceil(img_duration / C_sampl_sep))   # how many samples take in C to account for 30 seconds in x

# Movement of the window
move_window_each = 5 #seconds

samples_window_duration_C = np.ceil(img_duration / C_sampl_sep) # convert 30 seconds to samples in C
samples_window_stride_C = np.ceil(move_window_each / C_sampl_sep) # convert 15 seconds of sliding window in samples in C
```

```python
# Count the number of windows that fit in that problem
num_windows = int(np.ceil((C.shape[1] - samples_window_duration_C) / samples_window_stride_C))

 
for num_window in range(num_windows+ 1):
   
    idx_sample_window_start = int(np.ceil(num_window * samples_window_stride_C))
    idx_sample_window_end =  int(np.ceil(idx_sample_window_start + samples_window_duration_C))
    if idx_sample_window_end > C.shape[1]:
        idx_sample_window_end = C.shape[1] - 1
    print(f'Num Window: {num_window}; Start at: {idx_sample_window_start}; End at: {idx_sample_window_end}')
    
    # Take only windows that are complete
    matrix = SP_norm_flip[:,idx_sample_window_start:idx_sample_window_end]
    if matrix.shape[1] != C_samples_duration_window:
        continue
    print(matrix.shape)
```
